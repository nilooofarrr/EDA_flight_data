# Helpful Methods


```python
def capitalizer(df:pd.DataFrame,origin:str,destination:str,):
    origins = df.origin.str.capitalize()
    destinations = df.destination.str.capitalize()
    return (origins, destinations)

def date_list(df):
    return jdatetime.date(int(df[0:4]) ,int(df[5:7]) , int(df[8:10]) ).togregorian()

def str_to_list(input):
    change = {'[': '', ']': '','{' : '','.0': '', '}': '', '\'': '', ' ': '', 'nan':'0', '': '0'}
    output= ''
    for i , j in change.items():
        input = input.replace(i, j)
    output = input.split(',')
    output = [float(x) for x in output]
    return output

# str_to_list("[2041000.0, 2041000.0]nan")
```
